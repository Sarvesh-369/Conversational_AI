Extracting the contents of the Info box and appending to a list word by word 

```python
import requests
from bs4 import BeautifulSoup
import re
import html

# URL of the Wikipedia page with the infobox
url = 'https://en.wikipedia.org/wiki/Apple_Inc.'

# Send a GET request to the URL and retrieve the HTML content
response = requests.get(url)
html_content = response.text

# Create a BeautifulSoup object to parse the HTML
soup = BeautifulSoup(html_content, 'html.parser')

# Find the infobox element by its class or ID
infobox = soup.find('table', {'class': 'infobox'})

# Extract the key and words from the infobox
key_words = []
text=[]
rows = infobox.find_all('tr')
for row in rows:
    cells = row.find_all(['th', 'td'])
    if len(cells) >= 2:
        key = cells[0].get_text().strip()
        value = cells[1].get_text().strip()
        value = re.sub(r'\[.*?\]', '', value)  # Remove reference citations in square brackets
        value = re.sub(r'\n+', ' ', value)  # Replace multiple newline characters with a single space
        value = html.unescape(value)  # Decode HTML entities and escape sequences
        value = re.sub(r'\s+', ' ', value)  # Replace multiple consecutive spaces with a single space
        words = re.findall(r'\b\w+\b|\[.*?\]|\(.*?\)', value)  # Modified pattern to include parentheses and brackets
        key_words.append((key, words))

# Print the extracted key and words
for key, words in key_words:
    text.append(key)
    text.append(':')
    for word in words:
        text.append(word)
text
```
